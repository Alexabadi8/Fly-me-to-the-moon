```python
!pip3 install lxml
from bs4 import BeautifulSoup as bs
import requests as requests
import pandas as pd
import re
import numpy as np
import lxml as lx
import matplotlib as mpl
from datetime import date
%matplotlib inline
```

```python
#PROGRAM INSTRUCTIONS: Run all cells in order and consecutively for it to work correctly

#step 1
url = 'https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares.html'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

response = requests.get(url, headers=headers)
responseText = response.text
# Getting html content using beautiful soup
soup = bs(responseText, 'html.parser')
table = (soup.find("table", {"class":"table table-striped"}))
# Parsing text and finding proper table with data
prettyTable = table.prettify()
top = pd.read_html(prettyTable)
#Creating Pandas DataFrame
df = pd.concat(top)
#Renaming columns
new_column = ["rank","x_class", "date", "region", "start_time", "max_time" ,"end_time" ,"movie"]
df.columns = new_column
df

```

```python
#step 2

# Changing start_time, max_time, and end_time to include year (DateTime format)
df["start_time"] =  pd.to_datetime(df["date"] +" " + df["start_time"],format='%Y-%m-%d')
df["max_time"] =  pd.to_datetime(df["date"] +" " + df["max_time"],format='%Y-%m-%d')
df["end_time"] =  pd.to_datetime(df["date"] +" " + df["end_time"],format='%Y-%m-%d')


del df["date"]
df["movie"] = df["region"]
del df["region"]

# reassigning all headers
new_header = ["rank","x_class", "start_datetime", "max_datetime", "end_datetime", "region"]
df.columns=new_header


regex='\-+'
#Checking for non number values
df = df.replace(regex, value = np.nan, regex = True)

print(df.dtypes)
df
```

```python
#Step 3

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
nasa_url = "https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html"
r= requests.get(nasa_url,headers=headers)

# Getting html content using beautiful soup
soup = bs(r.content,features="html.parser")


#find html preformatted text
pre_tag = soup.find("pre")


# use regex to acquire start_date and start_time
regex1 = r"\d{4}\/\d{2}\/\d{2} \d{2}:\d{2} \d{2}\/\d{2} \d{2}:\d{2}"


# use regex to acquire flare location and flare region
regex2 = r"<\/a>\s*[a-zA-Z0-9\.\-\?]+\s*[a-zA-Z0-9\.\-\?]+\s{1}[a-zA-Z0-9\.\-\?]+\s{3}[<|\-]+|\s{3}[a-zA-Z0-9\.\-\?]+\s{3}" \
       r"[a-zA-Z0-9\.\-\?]+\s{1}[a-zA-Z0-9\.\-\?]+\s{1}[a-zA-Z0-9\.\-\?]+\s*[<|\-]" \
       r"|\s+[a-zA-Z0-9\.\-\?]+\s{3}[a-zA-Z0-9\.\-\?]+\s{1}[a-zA-Z0-9\.\-\?]+\s{1}[a-zA-Z0-9\.\-\?]+\s*[<|\-]"

# use regex to acquire frequencies, and all other values
regex3 = re.compile(r'(\d{4}/\d{2}/\d{2}\s{1,}\d{2}:\d{2}\s{1,}\d{2}/\d{2}\s{1,}\d{2}:\d{2}\s{1,}[\d\W]{4,}\s{1,}'
                     r'[\d\W]{2,}\s{1,}[A-Za-z\d\w\W]{4,7}\s{1,}[A-Za-z\d\w\W]{2,5}\s{1,}[A-Za-z\w\D\W]{1,4}\s{1,}'
                     r'[0-9\w\W]{2}/[0-9\w\W]{2}\s{1,}[0-9\w\W]{2}:[0-9\w\W]{2}\s{1,}[A-Za-z\d\w\W]{2,4}\s{1,}'
                     r'[A-Za-z\d\w\W]{2,4}\s{1,}[\d\w\W]{3,4}\s{1,}[\w\W]{4})')


grab_soup = soup.get_text()

match1 = re.findall(regex1,str(pre_tag))
match2 = re.findall(regex2,str(pre_tag))
match3 = regex3.findall(grab_soup)



splitter1 = []

splitter2 = []

splitter3 = []

for i in match1:
    splitter1.append(i.split())

for i in match2:
    splitter2.append(i.split())
for i in match3:
    splitter3.append(i.split())



# Creating individual columns for DataFrame

start_date = []
start_time = []
start_frequency = []
end_frequency= []
end_date = []
end_time =[]
cme_time = []
cpa = []
importance = []
flare_location = []
plot = []
width = []
speed = []
flare_region = []


for i in splitter1:
    start_date.append(i[0])
    start_time.append(i[1])
    end_date.append(i[2])
    end_time.append(i[3])

for i in splitter2:
    flare_location.append(i[1])
    flare_region.append(i[2])

for i in splitter3:
    start_frequency.append(i[4])
    end_frequency.append(i[5])
    importance.append(i[8])
    cme_time.append(i[10])
    cpa.append(i[11])
    width.append(i[12])
    speed.append(i[13])
    plot.append(i[14])

#Creating Matrix dataset
data = {
            "start_date": start_date,

           "start_time": start_time,

           "end_date": end_date,

           "end_time": end_time,

           "start_frequency": start_frequency,

           "end_frequency":end_frequency,

           "flare_location": flare_location,

            "flare_region": flare_region
}
#creating DF
df2=pd.DataFrame(data)
df2["importance"]= importance
df2["cme_datetime"] = cme_time

df2

```

```python
# Step 4

# Reformatting time values to DateTime format
df2["end_time"].iloc[0]

df2["end_time"] = np.where((df2.end_time == '24:00'),'00:00',df2.end_time) #changing 12am to 0

df2['end_date'] = df2['start_date'].str.split('/').str[0] + "/" + df2['end_date']

df2["end_date"] = pd.to_datetime(df2["end_date"] +" " + df2["end_time"],format='%Y-%m-%d')

df2["end_time"] = df2["start_date"]

df2["start_date"] = pd.to_datetime(df2["start_date"] +" " + df2["start_time"],format='%Y-%m-%d')


df2.rename(columns ={"start_date":"start_datetime"})

df2.rename(columns ={"end_date":"end_datetime"})


#cme time to datetime
df2["cme_datetime"] = pd.to_datetime(df2["end_time"] +" " + df2["cme_datetime"], errors='coerce',format='%Y-%m-%d')


#checking for NaN
regex4="\-+"
df2["cpa"]=cpa
df2= df2.replace(regex4, value = np.nan, regex = True)


df2.replace("Halo",np.nan,inplace=True) #replace halo with NaN
df2['width']=width
df2['speed']=speed


#if cpa == null -> True. False otherwise
df2['is_flare'] = np.where(df2.cpa.isnull(), True, False)
df2['plot']=plot
df2['width_lower_bound'] = np.where(df2.width.str.startswith(">"), True, False)




df2.rename(columns = {"start_date":"start_datetime"},inplace=True)


df2.rename(columns = {"end_date":"end_datetime"},inplace=True)


#dropping useless columns
df2.drop('start_time', axis=1, inplace=True)
df2.drop('end_time', axis=1, inplace=True)
df2.drop('plot', axis=1, inplace=True)

df2
```

```python
# Part 2

#Question 1 Replication


#first determine flares with X attached and separate in order to sort accordingly
topFlares= df2.loc[df2['importance'].str.startswith("X",na=False)].copy()

topFlares['importance'] = topFlares['importance'].astype('str')


topFlares['importance'] = topFlares['importance'].str.split("X").str.get(1)
#Since the the column is currently an object type we need to cast it to a float
topFlares['importance']= topFlares['importance'].astype('float64')
topFlares = topFlares.sort_values(by ='importance',ascending = False)

topFlares.index = np.arange(1, len(topFlares) + 1)

topFlares = topFlares.head(50)

topFlares['importance']= "X" + topFlares['importance'].astype('str')

#Reformatting to include x class after sorting
regex4="\-+"

topFlares= topFlares.replace(regex4, value = np.nan, regex = True)

#topFlares now contains the top 50 flare events from NASA

#When comparing the two datasets, it is obvious from first sight that they are similar but not the same.
#For example, NASA's set lowest flare importance is X1.9 when SWL's is X2.6
#Despite this, there are some rows which represent the same flare albiet with small imperfections
#See ranks 1-5 and notice how 4/5 show the same flares and how rank 3 says X17 in NASA but X17.2 in SWL

topFlares
```

```python
#Question 2 Integration

#This function takes in two datasets and checks for equivalence in two columns, start_datetime, and x_class
#Only reason I used these is because flare regions don't seem to match up

def bestMatchingRow(x, y):
    
    df = x.copy()
    df2 = y.copy()
    
    #Only using date because of slight differences between datasets
    df['start_datetime'] = df['start_datetime'].astype(str).str.split().str.get(0).astype(str)
    df2['start_datetime'] = df2['start_datetime'].astype(str).str.split().str.get(0).astype(str)

    df2['importance'] = df2['importance'].str.split('X').str[1].astype('float64')
    #separating X from importance
    df['x_class'] = df['x_class'].str.split('X').str[1].str.split('+').str[0].astype('float64')

    df2['rank']= np.where((df2['importance'].isin(df['x_class'])) & (df2['start_datetime'].isin(df['start_datetime']))
                 ,df['rank'], 'No Rank')

   
    return df2['rank']

topFlares['rank'] = bestMatchingRow(df,topFlares)

topFlares
```
